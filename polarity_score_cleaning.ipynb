```python
import pandas as pd
import numpy as np
print("done")
```

```python
df = pd.read_csv("text_results.csv")
```

```python
# subsetting the dirty data
uncleaned = df[['distillBert_tag','distillBert_ti']]

# extract the floats from text data
uncleaned['extract_tag'] = uncleaned.distillBert_tag.str.findall(r'(\d+(?:\.\d+)?)')
uncleaned['extract_ti'] = uncleaned.distillBert_ti.str.findall(r'(\d+(?:\.\d+)?)')

# expand into new columns
Neg_Pos_tag = pd.DataFrame(uncleaned['extract_tag'].tolist()).fillna('')
Neg_Pos_ti = pd.DataFrame(uncleaned['extract_ti'].tolist()).fillna('')

# rename columns
Neg_Pos_tag = Neg_Pos_tag.rename(columns={0:'neg_tag_disBert',1:'pos_tag_disBert'})
Neg_Pos_ti = Neg_Pos_ti.rename(columns={0:'neg_ti_disBert',1:'pos_ti_disBert'})

# merge columns
Neg_Pos_tag = Neg_Pos_tag.reset_index()
Neg_Pos_ti = Neg_Pos_ti.reset_index()
Neg_Pos_tag = Neg_Pos_tag.merge(Neg_Pos_ti,how='left')

# merge back to original dataFrame
df = df.reset_index()
df = df.merge(Neg_Pos_tag,how='left')

# rearranging columns
df = df[['tagline','neg_tag_vader','neu_tag_vader','pos_tag_vader','compound_tag_vader','neg_tag_disBert','pos_tag_disBert',
        'title','neg_ti_vader','neu_ti_vader','pos_ti_vader','compound_ti_vader','neg_ti_disBert','pos_ti_disBert']]

# convert to float64 datatypes
df['neg_tag_disBert'] = df.neg_tag_disBert.astype('float64')
df['pos_tag_disBert'] = df.pos_tag_disBert.astype('float64')
df['neg_ti_disBert'] = df.neg_ti_disBert.astype('float64')
df['pos_ti_disBert'] = df.pos_ti_disBert.astype('float64')

df.info()
```
