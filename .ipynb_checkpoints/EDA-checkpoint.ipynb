## ANOVA Test

<b>ANOVA test</b> is a statistical methods used in the testing of hypothesis for comparison of means among three or more groups. To perform it, we need to ensure that the following assumptions are to be met.
1) Sample data are randomly selected from populations and randomly assigned to each of the treatment groups. Each observation is thus independent of any other observation — randomness and independence.

2) Normality. Values in each sampled groups are assumed to be drawn from normally distributed populations. We can use normal probability plot or Q-Q plot to check normality. (Q-Q Plot)

3) Homogeneity of variance. All the c group variances are equal, that is σ₁² = σ₂² = σ₃² = … = σ𝒸². As a rule of thumb, if the ratio of the largest to the smallest sample standard deviation is less than 2, we consider the equal standard deviations assumption as fulfilled. (Homogenity Check)

References: 
1)https://www.ncbi.nlm.nih.gov/pmc/articles/PMC6813708/#:~:text=The%20Student's%20t%20test%20is,mean%20difference%20was%20statistically%20significant.
2) https://towardsdatascience.com/anova-test-with-python-cfbf4013328b

#### Normality Assumption Check

```python
def norm_check(x,y,data,figsize=(20,10),row=4):
    # creating temp data frame
    data_temp = data[[x,y]]
    
    # extracting no. of unique labels and initializing row and col
    unique_label = data[x].unique()
    len_unique = len(unique_label)
    row = row
    col = (len_unique//row) + 1
    count = 1
    
    # configuring graph size
    fig = plt.figure(figsize=figsize)
    plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=1.5,
                    hspace=1.5)
    
    # plotting to each subplot
    for i in unique_label:
        ax = plt.subplot(row,col,count)
        count+=1
        stats.probplot(data[data[x] == i][y], dist="norm", plot=ax)
        plt.title(i)
    plt.show()
    
```

#### Homogeneity of variance Assumption Check

```python
def homogen_check(x,y,data):
    data_temp = data[[x,y]]
    ratio = data_temp.groupby(x).std().max() / data_temp.groupby(x).std().min()
    print(ratio)
```

#### ANOVA Test

```python
def ANOVA(x,y,data):
    data_temp = data[[x,y]]
    CategoricalGroupLists = data_temp.groupby(x)[y].apply(list)
    AnnovaResults = f_oneway(*CategoricalGroupLists)
    
    # print results
    print("p-value:" + str(AnnovaResults[-1]))
    
    # null hypothesis: there is no difference among group means
    if AnnovaResults[1] > 0.05:
        print("We have insufficient evidence in rejecting the null hypothesis")
    else:
        print("We reject the null hypothesis")
```

### Relationship between feature and success state of campagin

We want to find out the relationship between 2 variable. To visualize, we will do a count plot of the 2 categorical variables first. Next, in order to evaluate the relationship, we will use Chi-Square Test and Cramer V to help us with intepretation of results, with level of significance of 5%
The coefficient of Cramer V suggest the following:
1) [0,0.2], fields are weakly associated

2) (0.2,0.6] fields are moderately associated

3) (0.6,1] fields are strongly associated

References:
1) https://towardsdatascience.com/levels-of-measurement-statistics-and-python-implementations-8ff8e7867d0b

2) https://www.ibm.com/docs/en/cognos-analytics/11.1.0?topic=terms-cramrs-v


```python
def cat_count(data,x,y):
    plt.figure(figsize=(16,6))
    arr = data[x].unique()
    arr = np.sort(arr)
    sns.countplot(data=data, x=x,palette='hls', hue=y, order=arr)
    plt.xticks(rotation=60)
    plt.show()
    display(pd.DataFrame(data[[x,y]].groupby([x,y])[y].count()))
```

```python
def cramerV(data,x,y):
    
    # obtaining a pivot table of count
    data_temp = data[[x,y]]
    data_temp['y_temp'] = data_temp[y].astype(str)
    data_temp = data_temp.groupby([x,y])['y_temp'].count().reset_index()
    data_temp = data_temp.pivot(index=y,columns=x)['y_temp'].reset_index().drop([y],axis=1)
    data_temp = data_temp.fillna(0.0)
    display(data_temp)
    
    # Chi-Square
    chiRes = stats.chi2_contingency(data_temp)
    print(f'chi-square statistic: {chiRes[0]}')
    print(f'p-value: {chiRes[1]}')
    if chiRes[1] > 0.05:
        print("We have insufficient evidence in rejecting the null hypothesis")
    else:
        print("We reject the null hypothesis")
    
    
    # sample size
    n = np.sum(np.sum(data_temp))
    minShape = min(data_temp.shape)-1
    chisq_stat = chiRes[0]
    V = np.sqrt( (chisq_stat/n)/minShape)
    print()
    print(f"Cramer' V: {V}")
```

## Relationship between campaign duration and amount of fund raised.

```python
# First, we will create a class for dates  
class date_n:  
    def __init__(self, day, month, year):  
        self.day = day  
        self.month = month  
        self.year = year  
   
   
# For storing number of days in all months from  
# January to December.  
month_Days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]  
   
# This function will count the number of leap years from 00/00/0000 to the #given date  
def count_Leap_Years(day):  
   
    years = day.year  
   
    # Now, it will check if the current year should be considered for the count          # of leap years or not.  
    if (day.month <= 2):  
        years -= 1  
   
    # The condition for an year is a leap year: if te year is a multiple of 4, and a            # multiple of 400 but not a multiple of 100.  
    return int(years / 4) - int(years / 100) + int(years / 400)  
   
    
# This function will return number of days between two given dates
def get_difference(date_1, date_2):  
   
    # Now, it will count total number of days before first date "date_1"  
   
    # Then, it will initialize the count by using years and day  
    n_1 = date_1.year * 365 + date_1.day  
   
    # then, it will add days for months in the given date  
    for K in range(0, date_1.month - 1):  
        n_1 += month_Days[K]  
   
    # As every leap year is of 366 days, we will add   
    # a day for every leap year  
    n_1 += count_Leap_Years(date_1)  
   
    # SIMILARLY, it will count total number of days before second date "date_2"  
   
    n_2 = date_2.year * 365 + date_2.day  
    for K in range(0, date_2.month - 1):  
        n_2 += month_Days[K]  
    n_2 += count_Leap_Years(date_2)  
   
    # Then, it will return the difference between two counts  
    return (n_2 - n_1)  
```

```python
df_time = df_new[['year_end', 'month_end', 'day_end', 'year_launch', 'month_launch', 'day_launch']]
df_amount_raised = df_new[['amount_raised_usd','amount_raised_usd_boxcox']]
```

```python
days = []
for i in range(0, 20618):
    year_end = df_time['year_end'].iloc[i]
    month_end = df_time['month_end'].iloc[i]
    day_end = df_time['day_end'].iloc[i]
    year_launch = df_time['year_launch'].iloc[i]
    month_launch = df_time['month_launch'].iloc[i]
    day_launch = df_time['day_launch'].iloc[i]
    
    date_1 = date_n(day_launch, month_launch, year_launch)  
    date_2 = date_n(day_end, month_end, year_end) 
    days.append(get_difference(date_1, date_2))


df_days = pd.DataFrame(days, columns = ["Total_Days"])
df_timenew = df_days.join(df_amount_raised)
```

```python
## Stat test algorithms

from scipy.stats import kendalltau
from scipy.stats import spearmanr
def pearson_corr(x,y,data):
    corr = data.corr()[y][x]
    print('Pearson correlation: %.5f' % corr)
    
def kendall_rank_corr(x,y,data):
    corr, _ = kendalltau(data[x], data[y])
    print('Kendall Rank correlation: %.5f' % corr)
def spearman_corr(x,y,data):
    rho, _ = spearmanr(data[x], data[y])
    print('Spearman\'s Correlation: %.5f' % rho)
    
def point_biserial_corr(x,y,data):
    result = stats.pointbiserialr(data[x], data[y])
    print(f'correlation between X and y: {result[0]:.2f}')
    print(f'p-value: {result[1]:.2g}')
    if result[1] > 0.05:
        print("We have insufficient evidence in rejecting the null hypothesis")
    else:
        print("We reject the null hypothesis")
```

```python
pearson_corr('compound_tag_vader','amount_raised_usd_boxcox',df_tag_res)
kendall_rank_corr('compound_tag_vader','amount_raised_usd_boxcox',df_tag_res)
spearman_corr('compound_tag_vader','amount_raised_usd_boxcox',df_tag_res)
```
